## Numpy Arrays
NumPy provides fast, efficient tools for numerical computation with arrays and matrices.

```python
# How to initialize an array?
A = np.arange(0, 10) #sequences of numbers
b = np.array([[1, 2, 3], [4, 5, 6]]) #standard arrays
c = np.ones(10) #arrays of 1s
d = np.zeros((2, 3)) #arrays of 0s
e = np.zeros(10, dtype=np.int32) #specify the type
A, b, c, d, e
```

### Operations on arrays

```python
A = np.arange(9).reshape((3, 3)) # reshape
A
```

```python
A = np.arange(0,100)
AC = A.reshape((10,10), order='C') #C-style ordering
AF = A.reshape((10,10), order='F') #F-style ordering
AC, AF
```

```python
a = np.random.random((2, 3))
print("a= ", a)
print(a.sum())
print(a.min())
print(a.max())
b = np.arange(12).reshape((3, 4))
print(b.min(axis=1))
a[0, 0] = 3 #access/change elements (1 element per axis!)
print("a= ", a)
```

```python
a = np.arange(10)**2
print(a[2:5])          # from index 2 to 5-1 (exclusive)
a[:6:2] = -3    # change elements from 0 to 5 every 2 with -3
print(a[::-1])
```

## Matplotlib
Matplotlib is a powerful plotting library that allows you to create a wide variety of static, animated, and interactive visualizations in Python.

```python
# Simple plot
x = np.linspace(0, 2, 100)
plt.plot(x, x, label='linear')
plt.xlabel('x label')
plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
# More advanced plot
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211) # add a subplot
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212) # add another subplot
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Scipy
SciPy builds on NumPy to offer advanced scientific functions such as optimization, integration, interpolation, and signal processing.

```python
from scipy.interpolate import interp1d

# Original data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Create interpolation functions (linear and cubic)
f_linear = interp1d(x, y, kind='linear')
f_cubic = interp1d(x, y, kind='cubic')

# New x values for interpolation
x_new = np.linspace(0, 5, 200)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'o', label='Data points')
plt.plot(x_new, f_linear(x_new), '-', label='Linear interpolation')
plt.plot(x_new, f_cubic(x_new), '--', label='Cubic interpolation')

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolation with scipy.interpolate.interp1d")
plt.show()
```
